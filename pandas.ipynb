```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 读取数据

```python
nba = pd.read_csv("./data/nba_2017_nba_players_with_salary.csv")
nba
```

# 数据处理

## 数据相关性

```python
data_corr = nba.loc[:,['RPM','AGE','SALARY_MILLIONS','ORB','TRB',
                       'AST','STL','BLK','TOV','PF','POINTS','GP',
                       'MPG','ORPM','DRPM']]
data_corr.head()
```

```python
# 获得各列之间的相关性
corr = data_corr.corr()
corr
```

对两个变量或多个变量之间相关关系的分析，叫做相关性分析。

通常用来分析两组或多组数据的变化趋势是否一致。

比如身高和体重是否存在关系，天气冷和袜子的销量是否存在关系，客户满意度和客户投诉率是否存在关系等。

```python
plt.figure(figsize=(20,8),dpi=80)
# 热力图
sns.heatmap(corr,square=True,linewidths=0.1,annot=True)
```

```python
# 用displot绘制球员薪水、效率值、年龄三个信息的分布情况
```

```python
# 设置背景
sns.set_style('darkgrid')
# 设置画布大小
plt.figure(figsize=(10,10))
# 画三幅图，这是第一行，第一列
plt.subplot(3,1,1)
sns.distplot(nba['SALARY_MILLIONS'])
plt.ylabel('salary')

plt.subplot(3,1,2)
sns.distplot(nba['RPM'])
plt.ylabel('rpm')

plt.subplot(3,1,3)
sns.distplot(nba['AGE'])
plt.ylabel('age')
```

### 多变量

### 衍生变量的可视化操作

```python
# 定义函数
def age_cut(df):
    if df.AGE <= 24:
        return 'young'
    elif df.AGE >= 30:
        return 'old'
    else:
        return 'best'
    
# 使用函数划分并添加到原数据中
nba['age_cut'] = nba.apply(lambda x:age_cut(x),axis = 1)
nba.head()
```

```python
nba['age']=1
```

```python
# 获得不同年龄段的球员薪水与热力值的关系
sns.set_style('darkgrid')
plt.figure(figsize=(8,8),dpi = 50)
plt.title('RPM AND SALARY',size = 15)
# 获得old和salary
x1 = nba.loc[nba.age_cut == 'old'].SALARY_MILLIONS
y1 = nba.loc[nba.age_cut == 'old'].RPM
plt.plot(x1,y1,'^')

x2 = nba.loc[nba.age_cut == 'young'].SALARY_MILLIONS
y2 = nba.loc[nba.age_cut == 'young'].RPM
plt.plot(x2,y2,'.')

x3 = nba.loc[nba.age_cut == 'best'].SALARY_MILLIONS
y3 = nba.loc[nba.age_cut == 'best'].RPM
plt.plot(x3,y3,'*')
```

```python
multi_data2 = nba.loc[:,['RPM','POINTS','TRB','AST','STL','BLK','age_cut']]
```

```python
# hue是标签值
sns.pairplot(multi_data2,hue = 'age_cut')
```

# 球队数据分析

## 球队薪资排行

```python
# agg函数
nba.groupby(by = 'age_cut').agg({'SALARY_MILLIONS':np.mean})
```

### 按球队分组，平均薪水，降序排序

```python
sal_team = nba.groupby(by = 'TEAM').agg({'SALARY_MILLIONS':np.mean})
```

```python
sal_team.sort_values(by='SALARY_MILLIONS',ascending = False)
```

### 按照球队分年龄段，上榜球员降序排列，上榜球员数相同，按效率值降序排列

```python
nba_rpm = nba.groupby(by = ['TEAM','age_cut']).agg({'RPM':np.mean,
                                                    'SALARY_MILLIONS':np.mean,
                                                    'PLAYER':np.size})
```

```python
nba_rpm
```

```python
nba_rpm.sort_values(by = ['PLAYER','RPM'],ascending = False)
```

### 球队综合实力排名

```python
nba_team = nba.groupby(by = 'TEAM',as_index = False).agg({'SALARY_MILLIONS':np.mean,
                                                         'RPM':np.mean,
                                                         'PLAYER':np.size,
                                                         'POINTS':np.mean,
                                                         'eFG%':np.mean,
                                                         'AGE':np.mean})
nba_team.head()
```

```python
nba_team.sort_values('RPM',ascending=False).head()
```

### 利用箱形图，小提琴图看10支球队的相关数据

```python
# 定义画布大小
sns.set_style('whitegrid')
plt.figure(figsize = (12,8))
nba_team1 = nba[nba['TEAM'].isin(['GS','CLE','SA','LAC','OKC','UTAH','CHA','TOR','NO','BOS'])]

#基于薪水
plt.subplot(3,1,1)
sns.boxplot(x='TEAM',y='SALARY_MILLIONS',data=nba_team1)

# 基于年龄
plt.subplot(3,1,2)
sns.boxplot(x='TEAM',y='AGE',data=nba_team1)

# 基于出场时间
plt.subplot(3,1,3)
sns.boxplot(x='TEAM',y='MPG',data=nba_team1)
```

### 利用小提琴图

```python
# 定义画布大小
sns.set_style('whitegrid')
plt.figure(figsize = (12,8))
nba_team1 = nba[nba['TEAM'].isin(['GS','CLE','SA','LAC','OKC','UTAH','CHA','TOR','NO','BOS'])]

#基于薪水
plt.subplot(3,1,1)
sns.violinplot(x='TEAM',y='SALARY_MILLIONS',data=nba_team1)

# 基于年龄
plt.subplot(3,1,2)
sns.violinplot(x='TEAM',y='AGE',data=nba_team1)

# 基于出场时间
plt.subplot(3,1,3)
sns.violinplot(x='TEAM',y='MPG',data=nba_team1)
```
